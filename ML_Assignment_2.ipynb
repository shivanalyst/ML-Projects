#**Machine Learning With Python: Linear Regression With One Variable**

**Problem Statement:** Predict canada's per capita income in year 2020. Using this build a regression model and predict the per capita income of canadian citizens in year 2020

**Steps to be followed for solving the Problem Statement:**

1.   Importing the Libraries
2.   Loading the Dataset
1.   Defining Dependent and Independent Variables
2.   Perform Exploratory data analysis
1.   Build a Linear Regression Model
2.   Predict the per capita income for 2020.







```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
# This command displays whether any NUll value is present or not
df.isnull( )
```

```python
# This command helps in understanding the size of the dataset
df.shape
```

```python
new_df = df.drop('income',axis='columns')
new_df
```

```python
Income = df.income
Income
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,Income)
```

```python
reg.predict([[2020]])
```
